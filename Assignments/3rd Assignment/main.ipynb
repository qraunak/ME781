##### Write a Python program that takes a 3D point as an input and calculates its Euclidean norm and Mahalanobis norm from the center of the two clusters of data given in the attachment. Visualize the clusters and the input data point in a 3D plot.

```python
import pandas as pd
import math
import numpy as np
```

```python
df1 = pd.read_csv('3D_cluster1.csv')
df1 = df1.to_numpy()
df1
```

```python
df2 = pd.read_csv('3D_cluster2.csv')
df2 = df2.to_numpy()
df2
```

```python
center1= np.mean(df1,axis=0)
center1
```

```python
center2= np.mean(df2,axis=0)
center2
```

```python
cov1 =np.cov(df1, rowvar=False)
cov1
```

```python
cov2 = np.cov(df2, rowvar=False)
cov2
```

```python
def euclidean_norm(point, center):
    x1, x2= point[0],center[0]
    y1, y2= point[1],center[1]
    z1, z2= point[2],center[2]
    distance = math.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)
    return distance
```

```python
def mahalanobis_norm(point, center, cov_matrix):
    diff = point - center
    inv_cov_matrix = np.linalg.inv(cov_matrix)
    return np.sqrt(np.dot(np.dot(diff.T, inv_cov_matrix), diff))
    
    
```

```python
x =float(input("Enter x Coordinate: "))
y =float(input("Enter y Coordinate: "))
z =float(input("Enter x Coordinate: "))
point =np.array([x,y,z])
```

```python
euclidean_norm1 = euclidean_norm(point, center1)
euclidean_norm2 = euclidean_norm(point, center2)
```

```python
mahalanobis_norm1 = mahalanobis_norm(point, center1, cov1)
mahalanobis_norm2 = mahalanobis_norm(point, center2, cov2)
```

```python

print(f"Euclidean Norm from Center of Cluster 1: {euclidean_norm1}")
print(f"Euclidean Norm from Center of Cluster 2: {euclidean_norm2}")
print(f"Mahalanobis Norm from Center of Cluster 1: {mahalanobis_norm1}")
print(f"Mahalanobis Norm from Center of Cluster 2: {mahalanobis_norm2}")
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df1[:, 0], df1[:, 1], df1[:, 2], c='blue', label='Cluster 1', alpha=1)
ax.scatter(df2[:, 0], df2[:, 1], df2[:, 2], c='green', label='Cluster 2', alpha=1)
ax.scatter(center1[0], center1[1], center1[2], c='red', marker='x', s=100, label='Center 1')
ax.scatter(center2[0], center2[1], center2[2], c='black', marker='x', s=100, label='Center 2')
ax.scatter(point[0], point[1], point[2], c='purple', marker='o', s=100, label='Input Point')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Visualization of Clusters and Input Point')

ax.legend()

plt.show()
```
